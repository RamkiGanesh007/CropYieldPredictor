```python
X = pd.DataFrame(df.iloc[:, 1:50].values)
y = df.iloc[:, 49].values
print(y)
```

```python

from sklearn.preprocessing import LabelEncoder
labelencoder_X_0 = LabelEncoder()
X.loc[:, 0] = labelencoder_X_0.fit_transform(X.iloc[:, 0])
labelencoder_X_1 = LabelEncoder()
X.loc[:, 1] = labelencoder_X_1.fit_transform(X.iloc[:, 1])
X
```

```python


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
X_train
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_squared_error
import seaborn as sns
 # create regressor object
model = RandomForestRegressor(n_estimators = 100, random_state = 0)
  
# fit the regressor with x and y data
model.fit(X, y)
print(X_test)
y_pred=model.predict(X_test)
# print(y_pred)
df=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
df
```

```python
# Save Model Using Pickle
import pandas
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
import pickle


# save the model to disk
filename = '/content/drive/MyDrive/mini/RF_model_Py3.pb'
pickle.dump(model, open(filename, 'wb'))
 
# some time later...
 
# load the model from disk
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(X_test, y_test)

print(result)



# with open('/content/drive/MyDrive/mini', 'wb') as f:
#   f.write(tflite_model)
```
